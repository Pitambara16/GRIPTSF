### Question : To Explore Unsupervised Machine Learning
                From the given ‘Iris’ dataset, predict the optimum number of
                clusters and represent it visually.

```python
x = df.iloc[:, [0, 1, 2, 3]].values
x
```

```python
sns.countplot(x='Species', data=df)
```

```python
from sklearn.cluster import KMeans
```

```python
#Elbow method to minimize WSS(Within-cluster Sum of Square)
Sum_of_squared_distances = []
Krange = range(1,11)
for k in Krange:
    km = KMeans(n_clusters=k)
    km = km.fit(x)
    Sum_of_squared_distances.append(km.inertia_)
    
```

```python
Sum_of_squared_distances
```

```python
plt.plot(Krange, Sum_of_squared_distances, 'bx-')
plt.xlabel('Krange')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow method for Optimal k')
plt.show()
```

```python
 #In the above plot, the elbow is at k=3 showing that the optimal k for the given data-set is 3.
```

```python
#Using K-Mean with k=3 for the data-set
km3 = KMeans(n_clusters=3, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
ykm3= km3.fit_predict(x)
ykm3
```

```python
print(km3.labels_)
```

```python
result = km3.labels_
result = pd.DataFrame(result,columns=['clusters'])
result.groupby('clusters').size()
```

```python
#Finding the centroids of the clusters
centroids = km3.cluster_centers_
centroids
```

```python
sns.set_style('darkgrid')
plt.style.use('seaborn')
plt.figure(figsize=[10,4])
plt.scatter(x[ykm3 == 0, 0], x[ykm3 == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[ykm3 == 1, 0], x[ykm3 == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[ykm3 == 2, 0], x[ykm3 == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(km3.cluster_centers_[:, 0], km3.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
